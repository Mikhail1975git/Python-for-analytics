##Задача2
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
fruits = {'apples': 5, 'bananas': 6, 'oranges': 4}

def vsego_fruits (dict):
    dict.values()
    count = 0
    for i in dict.values():
        count=count+i
    print ('total fruits = ', count)

vsego_fruits(fruits)
```

##Задача3
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]

Используйте list comprehensions

```python
ini_costs=[100, 125, -90, 345, 655, -1, 0, 200]

def total_costs(list):
    return {sum(i for i in list if i>0)}

total_costs(ini_costs)
```

###Задача 4.2

Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
purch_date =  ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
purch_amount =  [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

def get_income(purch_date: list, purch_amount: list) -> dict:
    income_monthly = {}
    for i in range(len(purch_date)):
        month = purch_date[i].split('-')[1]
        if month in income_monthly.keys():
            income_monthly[month] += purch_amount[i]
        else:
            income_monthly[month] = purch_amount[i]
    return income_monthly
get_income(purch_date, purch_amount)
```
