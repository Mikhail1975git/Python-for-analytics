# Задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x=df['waterfront'], y=df['price']/1000, whis=1.5)
plt.title("Зависимость цен от вида на набережную")
plt.xlabel("Вид на набережную")
plt.ylabel("Стоимость домов, тыс.ye");
```

Вывод: Как говорилось выше, только 1% домов имеет вид на набережную, и мы делали предположение, что это востребовано, престижно, и может вызвать повышенный спрос на данные дома, что в свою очередь, неизменно должно отразиться на стоимости. Данный анализ однозначно показыват, что наличие вида на набережную влияет на увеличение средней стоимости домов. Однако, мы должны отметить наличие "выбросов" в категории "без вида на набережную", которые говорят о наличии индивидуальных дорогостоящих предложений домов с уникальными или расширенными потребительскими характеристиками.

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x=df['grade'], y=df['price']/1000, whis=1.5)
plt.title("Зависимость цен от грэйда")
plt.xlabel("Грэйд")
plt.ylabel("Стоимость домов, тыс.ye");
```

Вывод: Данный график подтверждает логический вывод, что уровень Грэйда влияет на стоимость недвижимости. Однако мы видим что в интервале Грэйда от 3 до 6, фактическое влияние на стоимость минимальное.

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x=df['floors'], y=df['price']/1000, whis=1.5)
plt.title("Зависимость цен от этажности")
plt.xlabel("Количество этажей")
plt.ylabel("Стоимость домов, тыс.ye");
```

Вывод: Данный график показывает, что большинству потенциальных покапателей не очень важен такой показатель как - количество этажей. Большое количество "выбросов" говот о том, что приоритетными  по влиянию на стоимость, являются другие факторы.

```python
df['index_basement']=0
df.loc[df['sqft_basement'] > 0, "index_basement"] = 1
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['index_basement'], y=df['price']/1000, whis=1.5)
plt.title("Зависимость цен от наличия подвала")
plt.xlabel("Наличие подвала")
plt.ylabel("Стоимость домов, тыс.ye");
```

Вывод: Как и в предыдущем случае, данный график показывает, что потенциальным покапателям не очень важен такой показатель как - наличие подвала. Большое количество "выбросов" говот о том, что приоритетными повлиянию на стоимость, являются другие факторы.

```python
df.loc[df['sqft_basement'] <= 500, 'basement_group'] = 'Небольшие подвалы или нет'
df.loc[(df['sqft_basement'] > 500) & (df['sqft_basement'] <= 3000), 'basement_group'] = 'Средние подвалы'
df.loc[df['sqft_basement'] > 3000, 'basement_group'] = 'Большие подвалы'
```

```python
df.groupby('basement_group')['price'].mean().plot(kind='bar')
plt.xticks(rotation=10);
```

Вывод:Если в предыдущем анализе, мы увидели, что наличие подавала не является определяющим фактором влияющим на ценообразовании, то размер подвала определенно имеет такое влияние. Видимо таких домов не очень много.

```python
plt.figure(figsize=(20, 8))
sns.scatterplot(data=df, x="lat", y="long",  hue="price");
```

Вывод: данный точечный график визуализирует расположение дома по широте и долготе с хорошей четкостью. Скорее всего, больший уровень цен ( в интервале 47,6:47,7 градусов широты и (-122,2) градуса по долготе), отраженный более темным цветом, связан с близостью к ценному объекту, повышающему качество жизни, скорее всего озеру или реке.

```python
df.loc[df['sqft_living'] <= 1500, 'sqft_group'] = 'Небольшие дома'
df.loc[(df['sqft_living'] > 1500) & (df['sqft_living'] <= 3000), 'sqft_group'] = 'Средние дома'
df.loc[df['sqft_living'] > 3000, 'sqft_group'] = 'Большие дома'
```

```python
df.groupby('sqft_group')['price'].mean().plot(kind='bar')
plt.xticks(rotation=0);
```

Вывод: данный анализ подтверждает логическое заключение, что размер жилой площади напрямую влияет на стоимость дома

```python
df.loc[df['yr_built'] <= 1980, 'age_group'] = 'Старые дома'
df.loc[(df['yr_built'] > 1980) & (df['yr_built'] <= 2000), 'age_group'] = 'Ср. возраст домов'
df.loc[df['yr_built'] > 2000, 'age_group'] = 'Новые дома'
```

```python
df.groupby('age_group')['price'].mean().plot(kind='bar')
plt.xticks(rotation=0);
```

Вывод: на данном графике видно, что возраст дома в среднем не имеет определяющего влияния на уровень цен. Это можно объяснить, тем, что в соответствии с выводами по состоянию домов, большее количество домов находится в хорошем или очень хорошем состоянии вне зависимости от года постройки.

```python
df.drop(columns=['Unnamed: 0', 'id', 'sqft_lot', 'sqft_above', 'zipcode', 'sqft_living15', 'sqft_lot15', 'index_basement'], inplace=True)
```

```python
corr = df.corr()
import numpy as np
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

Вывод: Данная корреляция, обобщает сделанные выше выводы. Мы видим, что на стоимость домов оказывают влияние такие показатели, как площадь недвижимости(а соответственно количество спален и туалетных комнат), наличие вида на набережную, подтверждая это заключение зависимостью по географической широте в районе озера (или канала реки) и "Грэйд". Год постройки не сильно оказывает влияние на стоимость недвижимости, т.к. большинство домов находятся в хорошем или очень хорошем состоянии. Подвалами оборудовано небольшое количество домов, и это оказывает положительное влияние на цену дома, только если подвал имеет большие размеры. А количество этажей повлияет на цену, только если это в итоге дает большую площадь дома по сравнению с конкурентами.
